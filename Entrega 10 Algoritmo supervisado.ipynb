```python
import pandas as pd
import numpy as np 
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df= pd.read_csv("df_localidad.csv",  sep=',')
```

```python
df
```

```python
#borramos los datos NaN
df.dropna(inplace=True)
```

```python
df
```

```python
del df['Unnamed: 0']
```

```python
def Rendimiento(Rendimiento:str) -> int:
    """
    This function encodes the rendimiento variable.
    """
    
    if Rendimiento >= 2000: return 1
    else: return 0
```

```python
df["Rendimiento"] = df["Rendimiento"].apply(Rendimiento)
```

```python
df
```

```python
#cambiamos rendimiento por Target que es lo que buscamos predecir
df["target"] = df["Rendimiento"]
```

```python
df
```

```python
#borramos  la columna de Rendimiento que no la reemplazo "target"
del df["Rendimiento"]
```

```python
#borramos las columnas categoricas de "campa√±a" y "cultivo"
del df["Campa√±a"]
```

```python
del df["Cultivo"]
```

```python
del df["mes"]
```

```python
#Deep=true en el metodo Copy, signidica que quiero aplicar la data del DF originaly hacer cambios sin que esto afecte la data original
X = df.drop(columns=["target"]).copy(deep=True)
```

```python
X.head()
```

```python
X.info()
```

```python
y= df["target"].copy(deep=True)
```

```python
y
```

```python
#Preparando la data para el training
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
y_train
```

```python
#importamos algoritmos
from sklearn.neighbors import KNeighborsClassifier #KNN
from sklearn.linear_model import LogisticRegression #Regresion logistica
from sklearn.ensemble import RandomForestClassifier # Random Forest
```

```python
knn= KNeighborsClassifier() #neighbors = 5 (default)
lgr= LogisticRegression()   #(DEFAULT)
rfc= RandomForestClassifier() # trees=100 (Default)
```

```python
print(type(knn))
print(type(lgr))
print(type(rfc))
```

```python
#train Knn
knn.fit (X_train, y_train)
```

```python
#Predict whit KNN
knn.predict(X_test)
```

```python
y_test
```

```python
#train logistic Regretion
lgr.fit(X_train, y_train)
```

```python
#predict whit logistic Reg
lgr.predict(X_test)
```

```python
#train random forest
rfc.fit(X_train, y_train)
```

```python
#Predict whit random forest
rfc.predict(X_test)
```
