```python
#Importacion de librerias#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importacion de base de datos .csv#
df_estimacion = pd.read_csv("Estimaciones.csv", encoding='latin-1', sep=";")
df_estimacion
```

```python
#Informacion general del dataset#
df_estimacion.describe()
```

```python
#Detalles de columnas del dataset#
df_estimacion.columns
```

```python
#Seleccion de el valor a analizar#
girasol= df_estimacion.loc[:, 'Cultivo'] == 'Girasol'
df_girasol=df_estimacion.loc[girasol]
df_girasol
```

```python
#Grafico boxplot del rendimiento del cultivo de Girasol#
plt.figure(figsize=(10,10))
plt.boxplot(df_girasol["Rendimiento"])
plt.title("Rendimiento de Girasol", fontsize=20)
plt.xlabel('Kg/ha',fontsize=10)
plt.show()
```

```python
#Grafico de linea del rendimiento a traves del tiempo del cultivo de Girasol#
fig = sns.relplot(x = df_girasol["Campaña"], y = df_girasol["Rendimiento"], kind = "line",height=10, aspect=2)
plt.title("Rendimiento de Girasol", fontsize=20)
fig.set_xlabels("Campaña")
fig.set_ylabels("Kg/ha")
plt.show()
```

```python
#Grafico de barras del rendimiento a traves del tiempo del cultivo de Girasol#
x= df_girasol.Rendimiento
y= df_girasol.Campaña
df_girasol.values.flatten()
fig, ax = plt.subplots(figsize=(20,6))
ax.bar(y,x)
ax.set_title("Histograma Rendimiento", fontsize=20)
ax.set_xlabel('Campaña',fontsize=15)
ax.set_ylabel('Kg/ha',fontsize=15)
ax.grid(True)
```
