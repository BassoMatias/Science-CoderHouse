```python
#Se importan librerias json, requests y pandas#
import json
import requests
import pandas as pd

#Se obtienen datos de una URL#
url='https://jsonplaceholder.typicode.com/users'
```

```python
#Se establece la conexion y se prueba#
respuesta = requests.get(url)
print (respuesta)
```

```python
#Informacion en str#
respuesta.text
```

```python
#Lista de datos#
texto = respuesta.text
json.loads(texto)
```

```python
jsondata = json.loads(texto)
jsondata_flat = pd.json_normalize(jsondata)
```

```python
#Dataframe con pandas#
df = pd.DataFrame.from_dict(jsondata_flat)
df.head()
```
