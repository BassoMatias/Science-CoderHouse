```python
#Importacion de librerias#
import pandas as pd
import numpy as np
```

```python
#Lectura de los 2 archivos csv#
dfw= pd.read_csv ('w_data.csv', sep=';')
dfp= pd.read_csv ('p_data.csv', sep=';')
```

```python
#Visualizacion de archivos#
dfw
```

```python
#Visualizacion de archivos#
dfp
```

```python
#Fusion de DataFrame y visualizacion#
Fusionado= pd.merge(dfw, dfp, on='ID', indicator= 'true')
Fusionado
```

```python
#Reemplazar datos Target y chequear#
valores_remplazar = {' <=50K':'<=50K', ' <=50K.': '<=50K', ' >50K': '>50K', ' >50K.': '>50K'}

Fusionado.replace({'Target':valores_remplazar}, inplace = True)
Fusionado['Target'].unique()
```

```python
#Porcentaje segun Target y Sexo#
Mas50= Fusionado['Target']=='>50K'
Mas50K= Fusionado[Mas50]
PropS= Mas50K.groupby(['Target','Sex']).count()
PropS['Percent'] = (PropS['ID'] / 
                      PropS['ID'].sum()) * 100
PropS.Percent
```

```python
#Porcentaje segun Target y Raza#
PropR= Mas50K.groupby(['Target','Race']).count()
PropR['Percent'] = (PropR['ID'] / 
                      PropR['ID'].sum()) * 100
PropR.Percent
```

```python
#Porcentaje segun Target y Educacion#
PropE= Mas50K.groupby(['Target','Education']).count()
PropE['Percent'] = (PropE['ID'] / 
                      PropE['ID'].sum()) * 100
PropE.Percent
```
